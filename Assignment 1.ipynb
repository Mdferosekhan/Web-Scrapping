# Web Scrapping

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/laptops/pr?sid=6bo,b5g'
```

```python
res = requests.get(url)
```

```python
res
```

```python
wish = BeautifulSoup(res.text)
```

```python
wish
```

```python
Laptop = wish.select('._4rR01T')
```

```python
Laptop
```

```python
Laptop_title = []
```

```python
for item in Laptop:
    Laptop_title.append(item.text)
```

```python
Laptop_title
```

```python
Price = wish.select('._1_WHN1')
```

```python
Price
```

```python
release_Price =[]
```

```python
for item in Price:
    release_Price.append(item.text)
```

```python
release_Price
```

```python
Processor= wish.select('.rgWa7D:nth-child(1)')
```

```python
Processor
```

```python
release_processor = []
```

```python
for item in Processor:
    release_processor.append(item.text)
```

```python
release_processor
```
